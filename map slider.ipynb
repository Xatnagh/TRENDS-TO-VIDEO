```python
df_merged = pd.read_csv('CrimeStatebyState_1960-2014.csv')


	
### colorscale:    



	
scl = [[0.0, '#ffffff'],[0.2, '#ff9999'],[0.4, '#ff4d4d'], \
       [0.6, '#ff1a1a'],[0.8, '#cc0000'],[1.0, '#4d0000']] # reds



	
### create empty list for data object:    

data_slider = []
```

```python
for year in df_merged.Year.unique():


    # I select the year (and remove DC for now)
    df_sected_crime = df_merged[(df_merged['State']!= 'District of Columbia' ) &  (df_merged['Year']== year )]

    for col in df_sected_crime.columns:  # I transform the columns into string type so I can:
        df_sected_crime[col] = df_sected_crime[col].astype(str)

    ### I create the text for mouse-hover for each state, for the current year    
    df_sected_crime['text'] = df_sected_crime['State'] + ' Pop: ' + df_sected_crime['Population'] + ' Murder rate: ' + df_sected_crime['Murder_per100000']

    ### create the dictionary with the data for the current year
    data_one_year = dict(
                        type='choropleth',
                        locations = df_sected_crime['State_code'],
                        z=df_sected_crime['Murder_per100000'].astype(float),
                        locationmode='USA-states',
                        colorscale = scl,
                        text = df_sected_crime['text'],
                        )

    data_slider.append(data_one_year)  # I add the dictionary to the list of dictionaries for the slider
```

```python
steps = []

for i in range(len(data_slider)):
    step = dict(method='restyle',
                args=['visible', [False] * len(data_slider)],
                label='Year {}'.format(i + 1960)) # label to be displayed for each step (year)
    step['args'][1][i] = True
    steps.append(step)


	
##  I create the 'sliders' object from the 'steps' 

sliders = [dict(active=0, pad={"t": 1}, steps=steps)]  
```

```python
layout = dict(geo=dict(scope='usa',
                       projection={'type': 'albers usa'}),
              sliders=sliders)


	
# I create the figure object:

fig = dict(data=data_slider, layout=layout) 



	
# to plot in the notebook

plotly.offline.iplot(fig)



	
# to plot in a separete browser window

```

```python
from pytrends.request import TrendReq
pytrends = TrendReq(hl='en-US', tz=360)
import us
import country_converter as coco
```

```python
kw_list=['Taylor Swift']
pytrends.build_payload(kw_list, cat=0, timeframe='2016-12-14 2017-01-25', geo='', gprop='')

df = pytrends.interest_over_time()

datestampsss = []

for row in df.index:
    datestampsss.append((str(row)[:10]))
nf = pytrends.interest_by_region()
nf['geo'] = nf.index
nf['code'] = coco.convert(names=nf['geo'], to='ISO3')
nf
```

```python
i=0

for date in datestampsss:
    datestamp = date+" "+date
    pytrends.build_payload(kw_list, cat=0, timeframe=datestamp, geo='', gprop='')
    df = pytrends.interest_by_region()
```

```python

breeh = ['2016-12-14', '2016-12-15', '2016-12-16', '2016-12-17', '2016-12-18']

dsliders = []

for date in breeh:
    brih = date+" "+date
    pytrends.build_payload(kw_list, cat=0, timeframe=brih, geo='', gprop='')
    df = pytrends.interest_by_region()
    df['geo'] = df.index
    df['code'] = coco.convert(names=nf['geo'], to='ISO3')
    
    dmap = dict(
                type='choropleth',
                locations = df['code'],
                z=df[kw_list[0]].astype(float),
                locationmode='ISO-3',
                colorscale = 'Reds',
                colorbar_title = "Poopularity",
                )
    
    dsliders.append(dmap)
   
    
steps = []

for i in range(len(dsliders)):
    step = dict(method='restyle',
                args=['visible', [False] * len(dsliders)],
                label='Year {}'.format(breeh[i])) # label to be displayed for each step (year)
    step['args'][1][i] = True
    steps.append(step)



sliders = [dict(active=0, pad={"t": 1}, steps=steps)]  

layout = dict(title= "Popularity Over Time Map for " + kw_list[0],
              geo=dict(showframe=False,
                       showcoastlines=False,
                       projection={'type': 'equirectangular'}),
              sliders=sliders)


	
# I create the figure object:

fig = dict(data=dsliders, layout=layout) 



	
# to plot in the notebook

plotly.offline.iplot(fig)


```
